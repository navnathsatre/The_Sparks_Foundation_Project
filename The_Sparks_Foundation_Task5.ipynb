# Visualizing the datasets

## Number of matches in each city

```python
plt.style.use('classic')
plt.figure(figsize=(18,8))
sns.barplot(x=merge_data['city'].value_counts()[:10].index, y=merge_data['city'].value_counts()[:10].values)
plt.xlabel('City',size=20);plt.ylabel('No. of matches',size=20);plt.xticks(size = 18);plt.yticks(size = 18)
plt.title('No. of Matches in diff. cities',size=25,fontweight="bold")
plt.show()
```

### *Most of the matches are happened in "Mumbai" i. e (24302)*

## Number of matches played in each season

```python
plt.figure(figsize=(18,6))
sns.countplot('season', data = matches,palette='PiYG')
plt.title("Number of matches played in each season",fontsize=25,fontweight="bold")
plt.ylabel("Count", size = 20);plt.xlabel("Season", size = 20)
plt.xticks(size = 18);plt.yticks(size = 18)
plt.show()
```

### *2011-2013 have more matches being played than other seasons, All other seasons have approximately 58-60 matches while 2011-2013 have more than 70 matches.*

## How many teams played in each season?

```python
matches.groupby('season')['team1'].nunique().plot(kind = 'bar', figsize=(18,6), color='orange')
plt.title("Number of teams participated each season ",fontsize=25,fontweight="bold")
plt.ylabel("Count of teams", size = 20);plt.xlabel("Season", size = 20);plt.xticks(size = 18);plt.yticks(size = 18)
plt.show()
```

### *10 teams played in 2011 and 9 teams each in 2012 and 2013*

## Number of matches played at diffirent venue

```python
plt.figure(figsize=(18,10))
sns.barplot(y=merge_data['venue'].value_counts()[:10].index, x=merge_data['venue'].value_counts()[:10].values)
plt.xlabel('No. of matches',size=25);plt.ylabel('Venue',size=25);plt.xticks(size = 18);plt.yticks(size = 18)
plt.title("Most number of IPL matches at venue",fontsize=35,fontweight="bold")
plt.show()
```

### *Eden Gardens at Kolkatahas hosted the highest number of matches so far in IPL*

## Maximum wins in IPL

```python
winning_teams = matches[['season','winner']]
#dictionaries to get winners to each season
winners_team = {}
for i in sorted(winning_teams.season.unique()):
    winners_team[i] = winning_teams[winning_teams.season == i]['winner'].tail(1).values[0]    
winners_of_IPL = pd.Series(winners_team)
winners_of_IPL = pd.DataFrame(winners_of_IPL, columns=['team'])
winners_of_IPL['team'].value_counts().plot(kind = 'barh', figsize = (18,8), color = 'darkblue')
plt.title("Winners of IPL across 11 seasons",fontsize=25,fontweight="bold")
plt.ylabel("Teams", size = 20);plt.xlabel("Frequency", size = 20)
plt.xticks(size = 15);plt.yticks(size = 15);plt.plot()
```

### *'Mumbai Indians' wins IPL 4 times, 'Chennai Super Kings'  3 times and 'Gujarat Lions','Rising Pune Supergiant','Royal Challengers Bangalore','Delhi Daredevils', 'Kings XI Punjab' NEVER win IPL Trophies*

## Does teams choosed to bat or field first, after winning toss?

```python
matches['toss_decision'].value_counts().plot(kind='pie', fontsize=14, autopct='%3.1f%%', figsize=(10,6), shadow=True,
                                             startangle=90, legend=True, cmap='tab20')
plt.title('Decision taken by captains after winning tosses', size=25);plt.ylabel('Toss Decision');plt.show()
```

###  *61.2% times teams who have won tosses have decided to chase down*

## How toss decision affects match results?

```python
plt.figure(figsize = (18,6))
matches['toss_win_game_win'] = np.where((matches.toss_winner == matches.winner),'Yes','No')
sns.countplot('toss_win_game_win', data=matches, hue = 'toss_decision')
plt.title("How Toss Decision affects match result", fontsize=25,fontweight="bold")
plt.xticks(size = 25);plt.yticks(size = 15)
plt.xlabel("Winning Toss and winning match", fontsize = 20);plt.ylabel("Frequency", fontsize = 20);plt.show()
```

### *Teams winning tosses and electng to field first have won most number of times.*

## Individual teams decision to choose bat or field after winning toss

```python
plt.figure(figsize = (16,6))
sns.countplot('toss_winner', data = matches, hue = 'toss_decision',palette='mako')
plt.title("Teams decision to bat first or second after winning toss", size = 25, fontweight = 'bold')
plt.xticks(size = 10,rotation=25);plt.yticks(size = 15)
plt.xlabel("Toss Winner", size = 20);plt.ylabel("Count", size = 20);plt.show()
```

### *Most teams field first after winning toss except for Chennai Super Kings who has mostly opted to bat first. Deccan Chargers and Pune Warriors also show the same trend.*

## Teams total scoring runs, over the years

```python
merge_data.groupby('season')['batsman_runs'].sum().plot(linewidth = 3, figsize =(15,5),color = 'red')
plt.title("Runs over the years",fontsize= 25, fontweight = 'bold')
plt.xlabel("Season", size = 20)
plt.ylabel("Total Runs Scored", size = 20)
plt.xticks(size = 12);plt.yticks(size = 12);plt.show()
```

### *Run scoring has gone up from the start of the IPL in 2008*

## Top Run Getters of IPL

```python
merge_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'red',
                                                                                                 figsize = (15,5))
plt.title("Top Run Getters of IPL", fontsize = 20, fontweight = 'bold')
plt.xlabel("Batsmen", size = 20);plt.ylabel("Total Runs Scored", size = 20)
plt.xticks(size = 12);plt.yticks(size = 12);plt.show()
```

### Suresh Raina is the highest run getter in IPL, Except for MS Dhoni, all other top run getters are either openers or come in 3rd or 4th positions to bat

## Most Wicket taker bowler

```python
merge_data.groupby('bowler')['player_dismissed'].count().sort_values(ascending = False).head(10).plot(kind = 'bar', 
                                                color = 'purple', figsize = (15,5))
plt.title("Top Wicket Takers of IPL", fontsize = 20, fontweight = 'bold')
plt.xlabel("Bowler", size = 20);plt.ylabel("Total Wickets Taken", size = 20);plt.xticks(size = 12);plt.yticks(size = 12);plt.show()
```

### *Malinga has taken the most number of wickets in IPL followed by Bravo and Amit Mishra*

## Which batsman has been most consistent among top 10 run getters

```python
consistent_batsman = merge_data[merge_data.batsman.isin(['SK Raina', 'V Kohli','RG Sharma','G Gambhir','RV Uthappa', 'S Dhawan',
                                                         'CH Gayle', 'MS Dhoni','DA Warner', 'AB de Villiers'])][['batsman','season',
                                                                                                                  'total_runs']]
consistent_batsman.groupby(['season','batsman'])['total_runs'].sum().unstack().plot(kind = 'box', figsize = (15,6))
plt.title("Most Consistent batsmen of IPL", fontsize = 25, fontweight = 'bold')
plt.xlabel("Batsmen", size = 20);plt.ylabel("Total Runs Scored each season", size = 20);plt.xticks(size = 15,rotation=20)
plt.yticks(size = 15);plt.show()
```

### *Median score for Raina is above all the top 10 run getters. He has the highest lowest run among all the batsmen across 11 seasons. Considering the highest and lowest season totals and spread of runs, it seems Raina has been most consistent among all.*

## Batsmen with the best strike rates over the years 

```python
#We will consider players who have played 10 or more seasons
no_of_balls = pd.DataFrame(merge_data.groupby('batsman')['ball'].count()) #total number of matches played by each batsman
runs = pd.DataFrame(merge_data.groupby('batsman')['batsman_runs'].sum()) #total runs of each batsman
seasons = pd.DataFrame(merge_data.groupby('batsman')['season'].nunique()) #season = 1 implies played only 1 season

batsman_strike_rate = pd.DataFrame({'balls':no_of_balls['ball'],'run':runs['batsman_runs'],'season':seasons['season']})
batsman_strike_rate.reset_index(inplace = True)

batsman_strike_rate['strike_rate'] = batsman_strike_rate['run']/batsman_strike_rate['balls']*100
highest_strike_rate = batsman_strike_rate[batsman_strike_rate.season.isin([10,11])][['season',
                                          'batsman','strike_rate']].sort_values(by = 'strike_rate',ascending = False)
highest_strike_rate.head(10)
```
